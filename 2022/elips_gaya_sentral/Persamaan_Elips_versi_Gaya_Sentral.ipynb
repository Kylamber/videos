Jangan kabur dulu habis ngelihat ini, gambarnya menarik banget. 
1. Kalau kita lihat baik-baik, ada dua titik yang akan kita namakan f dan f aksen. Kedua titik ini bagaikan pusar lingkaran. Kalau kita tarik garis dari f ke suatu titik di keliling elipsnya, kita namakan r, dan lakukan juga yang sama dengan f aksen, kita namakan r aksen, jumlah r dan r aksen ini akan selalu sama. Mirip kan sama pusat lingkaran yang jaraknya selalu sama ke kelilingnya.

2. Sekarang kita coba lihat yang f, r, sama sudut theta aja. Kalau sudutnya 0, kita dapatkan r terpendek yang punya nama spesial, yaitu periapsis. Kalau sudutnya 180 derajat, kita dapatkan r terpanjang yang namanya adalah apoapsis. Kalau sedang ngebahas tentang orbit, ini bakal muncul. Ada nama lainnya juga, perigee dan apogee kalau ngebahas tentang orbit suatu benda dengan bumi atau perilune dan apolune kalau dengan bulan. Peri yang dekat, apo yang jauh. 

3. Lanjut lagi, pernah dengar kalau bicara tentang orbit itu ada yang namanya semimajor axis sama semiminor axis? Itu adalah a dan b dalam gambar. semimajor, seperti kata majornya, adalah jarak terpanjang dari titik tengah elips ke kelilingnya, berarti yang a. Sebaliknya, semiminor itu yang terpendek, jadi yang b.

4. Nah, kita masuk ke yang seru nih, eccentricity. Kalau diterjemahkan ke Bahasa Indonesia, itu artinya keanehan. Agak makes sense kalau kita lihat apa eccentricity dalam gambar ini. Kalau kita lihat eccentricity dikalikan dengan a, hasilnya itu adalah jarak fokus ke titik tengah elips. Gimana menurut kalian kalau eccentricitynya 0? Kedua fokusnya akan berkumpul di titik tengah elips dan hasilnya adalah... lingkaran? Selamat bagi yang sudah nyadar, eccentricity ini bisa dibilang sebagai "seberapa lonjong" elipsnya. Apa hubungannya sama terjemahan tadi yang artinya keanehan? Cuma teoriku doang sih, jadi kalau elips itu lingkaran yang aneh, jadi semakin lonjong semakin aneh. 

Aku bakal ngebahas lebih lanjut tentang eccentricity ini karena sangat menarik, kalau dia 0 tadi kita sudah tau dia akan membentuk lingkaran. Apa berarti yang di atas 0 itu akan selalu membentuk elips? Kita coba lihat rumus elipsnya

$$r = \frac{a(1-\epsilon^2)}{1+\epsilon\cos\theta}$$

a itu tadi semimajornya, epsilon itu eccentricitynya, dan theta itu sudutnya tadi. $a(1-\epsilon^2)$ bisa kita misalin aja jadi $\alpha$.

Lihat kalau kita masukkan 0 untuk eccentricitynya, radiusnya selalu a

$$r = a$$

Makes sense, karena dia lingkaran bukan?

Gimana kalau eccentricitynya 1?

$$r = 0$$

tiba-tiba radiusnya jadi 0... Penjelasan untuk ini adalah persamaan tadi cuma valid untuk epsilon dari 0 sampai kurang dari 1. Gimana kalau yang lebih dari itu?

eccentricity 1 itu berarti parabola dan lebih dari itu adalah hiperbola. Sedikit fun fact, [eccentricity dari bumi](https://nssdc.gsfc.nasa.gov/planetary/factsheet/earthfact.html) itu cuma 0.0167, jadi lebih keliatan kayak mengorbit seperti lingkaran daripada elips. Buat komet, eccentricitynya lumayan besar, misa [komet Halley](https://nssdc.gsfc.nasa.gov/planetary/factsheet/cometfact.html) eccentricitynya 0.967 yang berarti sangat lonjong.

Sekarang sudah tau anatomi dari elips, mungkin kita sedikit meluncur ke fisikanya. 

1. Dalam gaya sentral, momentum angular yang dialami benda itu selalu konstan. Apa itu momentum angular? Secara sederhana itu adalah massa dikalikan radius dan dilakukan kecepatan. Ini mengapa kalau suatu benda dalam orbit, ketika jauh dari benda orbitnya dia melambat dan ketika di jarak terdekannya dia mencepat. Radius yang makin besar? Bayarannya adalah kecepatannya. Kecepatan semakin besar? Bayarannya adalah radiusnya.

2. Kalau masih inget di dalam video aku menceritakan tentang Edelbaum Escape, di situ ada manuver di mana roketnya ganti orbit. Kalau kita mendorong roket dalam waktu yang singkat ke kecepatan tertentu, orbitnya akan berubah dengan lokasi roket mempercepat dirinya itu adalah periapsisnya.

Kita punya persamaan untuk mengetahui periapsis atau apoapsis dari suatu orbit
$$r_0 = \frac{\alpha}{1+\epsilon}\,\mathrm{(periapsis)}$$
$$r_1 = \frac{\alpha}{1-\epsilon}\,\mathrm{(apoapsis)}$$

Kita akan pakai persamaan periapsis karena disitu tempat kita mengubah kecepatan kita. 

$$r_0 = \frac{\alpha}{1+\epsilon}$$
$$\epsilon = \frac{\alpha}{r_0} - 1$$
$$\epsilon = \frac{ml^2}{kr_0} - 1$$
$$\epsilon = \frac{mr_0v^2}{k} - 1$$
$$\epsilon = \frac{r_0v^2}{GM} - 1$$

Setelah dapat persamaan di atas, kita bisa langsung aja pakai persamaan di atas untuk tau dengan kecepatan $v$ yang baru, eccentricity dari orbitnya berapa. Kayaknya cukup untuk saat ini, semoga berguna, bhay!
